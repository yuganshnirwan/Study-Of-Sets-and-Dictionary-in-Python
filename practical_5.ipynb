```python
# Creation of Set
stationary={"pencil","pen","scale"}
print(stationary)
```

```python
# Duplicates not allowed
stationary={"pencil","pen","scale","pen"}
print(stationary)
```

```python
# True and 1 are considered same
stationary={"pencil","pen","scale",True, 1, False, 0}
print(stationary)
```

```python
# Set wiht strings, integers and boolean values
thisset={'Yugansh',38,False,18.8}
print(thisset)
```

```python
#Searching in Set
stationary={"pencil","pen","scale"}
print('pencil' in stationary)
print('eraser' in stationary)
```

```python
# Items in sets are not replaceable, but items can be added and deleted from them
stationary={"pencil","pen","scale","pen"}
stationary.add('eraser')
stationary.remove('pen')

print(stationary)
```

```python
# Set operations
a={1,2,6,4}
b={5,4,7,6}

print('Union:', a|b)
print('Intersection:',a & b)
print('Difference:',a-b)
print('Symmetric Difference:', a^b)
```

```python
# Frozen Set
f_s=frozenset({"pencil","pen","scale"})
print(f_s)
print(type(f_s))
```

```python
# Adding and removing is not allowed in frozen set
f_s=frozenset({"pencil","pen","scale"})
f_s.add('eraser')
f_s.remove('pen')

print(f_s)
```

#Dictionary

```python
# Creation of Dictionary
thisdict={
    'company':'apple',
    'product':'iphone',
    'year': 2025
}
print(thisdict)
```

```python
# Dictionary takes latest value given to key. It doesnt allow duplicates
thisdict={
    'company':'apple',
    'product':'iphone',
    'year': 2023,
    'year':2025
}
print(thisdict)
```

```python
# Length of key-value pair
len(thisdict)
```

```python
# Changing/ Adding key:value in dictionary
product={
    'company':'apple',
    'product':'iphone',
    'year': 2025
}
product['colour']='lavender'
x=product.values()
y=product.keys()
print(x)
print(y)
```

```python
customer={
    'name':'Ayush',
    'product':'Charger',
    'date':'30 jan'
}
print('Customer Name:',customer['name'])
```

```python
# Adding and Updating Elements
customer['product2']='Headphones'
customer['name']='Prek'
print(customer)
```

```python
# Removing Elements
customer.pop('date')
print('updated details:',customer)
```
